```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("data3.csv")
```

```python
data.head(3)
```

## Get frequency of distribution

```python
# method 1
plt.hist(data["price"]) 
plt.show()
```

```python
# set width yourself by using bins
plt.hist(data["price"], bins=20)
plt.show()
```

```python
# cumulative frequency distribution
plt.hist(data["price"], bins=20, cumulative=True)
plt.show()
```

```python
# cumulative proportion distribution
plt.hist(data["price"], bins=10, normed=1)
plt.show()
```

```python
plt.hist(data["price"], bins=10, normed=1, cumulative=True)
plt.show()
```

```python
# method 2
sns.distplot(data["price"])
```
